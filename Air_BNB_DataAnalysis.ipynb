#Handling Missing Values

```python
df['last review'] = pd.to_datetime(df['last review'], errors='coerce')

df.fillna({'review per month': 0,'last review': df['last review'].mean()}, inplace=True)

df.dropna(subset = ['NAME', 'host name'], inplace=True)
```

```python
print(df.isnull().sum())
```

#Correct Data Types

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

#Remove Duplicates

```python
df.drop_duplicates(inplace=True)
```

#Confirm Data Cleaning

```python
print(df.info())
```

```python
df = df.drop(columns=['license','house_rules'], errors='ignore')
```

#Visualization

#Distribution of Prices

plot the distribution of listing prices

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True, color = 'red')
plt.title('Distribution of Listing Prices')
plt.xlabel('Price ($)')
plt.ylabel('Frequency')
plt.show()
```

#Room Type Analysis

Analyze the distribution of different room types

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='room type', color = 'hotpink')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

# Neighbourhood Analysis

Examine how listing are distributed across different neighborhood

```python
plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='neighbourhood group', color = 'lightgreen', order=df['neighbourhood group'].value_counts().index)
plt.title('Neighbourhood Distribution')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

#Price vs. Room Type

Visualize the relationship between price and room type

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='room type', y='price', hue = 'room type', color = 'lightblue', palette = 'Set1')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

#Review Overtime

Plot the number of reviews overtime

```python
df['last review']= pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', color = 'blue')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
